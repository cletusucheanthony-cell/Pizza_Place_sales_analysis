## Merging Data

```python
order_details = pd.merge(order_details, order, on='order_id', how='inner')
order_details
```

```python
pizzas = pd.merge(pizza_types, pizzas, on='pizza_type_id', how='inner')
pizzas
```

```python
final_merg = pd.merge(order_details, pizzas, on='pizza_id')
final_merg
```

```python
final_merg.isnull().sum()
```

```python
final_merg.shape
```

```python
final_merg.describe()# summarising the statistical form of the merged dataset
```

```python
final_merg['Month'] = pd.to_datetime(final_merg['date'], infer_datetime_format=True).dt.month #creating a column for Month only
final_merg['Hours'] = pd.to_datetime(final_merg['time'], infer_datetime_format=True).dt.hour#creating a column for Hours only
```

```python
final_merg.info()
```

## 1. Total Revenue/Sales

```python
final_merg.columns
```

```python
final_merg.info()
```

```python
final_merg['revenue'] = final_merg['quantity'] * final_merg['price']
```

```python
final_merg['revenue']
```

```python
total_revenue = final_merg['revenue'].sum()#calculating the total revenue generated over the years
total_revenue
```

```python
#plotting a graph of the total revenue per year
total_revenue_per_year = final_merg.groupby(final_merg['date'].dt.year)['revenue'].sum().reset_index()
total_revenue_per_year
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="date", y="revenue", data=total_revenue_per_year, palette="viridis")
plt.title("Total Revenue Performance Per Year", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Revenue($)", fontsize=12)

plt.savefig('total_revenues.png', dpi=300, bbox_inches='tight')

plt.show()
```

## Total Quantity

```python
total_qty = final_merg["quantity"].sum()
total_qty
```

```python
final_merg.head()
```

```python
# determing and visualize quantity by month
final_merg['Month'] = final_merg['date'].dt.month_name()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
qty_by_month = final_merg.groupby('Month')['quantity'].sum().reindex(month_order).reset_index()
qty_by_month
```

```python
# Monthly quantity trend
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='quantity', data=qty_by_month, marker='o', color='crimson', linewidth=3)
plt.fill_between(qty_by_month["Month"], qty_by_month["quantity"], color='gold', alpha=0.1)
plt.title("Monthly Sales Volume(Quantity)", fontsize=15, fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel("Pizzas Sold")

plt.savefig("qty_by_month.png", dpi=300)

plt.show()
```

```python
# Determing and visualize Quantity by category
qty_by_category = final_merg.groupby("category")["quantity"].sum().sort_values(ascending=False).reset_index()
qty_by_category
```

```python
# Category Quantity trend
plt.figure(figsize=(10, 6))
sns.barplot(x="category", y="quantity", data=qty_by_category, palette="flare")
plt.title("Total Quantity Sold by Pizza Category", fontsize=15, fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Pizzas Sold")

plt.savefig("qty_category.png", dpi=300, bbox_inches='tight')

plt.show()
```

## Average Price of Pizzas

## Peak Hours Of Sales

## Sales Per Day Of The Week

```python
# Calculating sales per day of the week
final_merg["day_name"] = final_merg['date'].dt.day_name()
final_merg["day_name"]
```

```python
days_order = ['Monday', 'Tuesday', 'Wenesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
days_order
```

```python
sales_per_day = final_merg.groupby("day_name",)["revenue"].sum().reindex(days_order).reset_index()
sales_per_day
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="day_name", y="revenue", data=sales_per_day, palette='viridis')
plt.title("Sales Per Day In a Week", fontsize=14, fontweight='bold')
plt.xlabel("Days In a Week", fontsize=12)
plt.ylabel("Total Revenue")

plt.savefig("sales_per_day.png", dpi=300, bbox_inches='tight')
plt.show()
```
